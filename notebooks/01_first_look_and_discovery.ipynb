## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

```python
# DOES CONTENT AGE DICTATE TREND DIRECTION?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the data is loaded
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Calculate descriptive statistics of content age (in days) across trend directions
age_stats = df.groupby("trend_direction")["content_age_days"].agg(["count", "median", "mean", "std"])
print("--- Content Age (Days) Metrics by Trend Direction ---")
print(age_stats.round(1).to_string())

# Visualize the distribution of content age across performance trends
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df,
    x="trend_direction",
    y="content_age_days",
    order=["down", "flat", "stable", "up", "new"],
    palette="muted"
)
plt.title("Distribution of Content Age (Days) by Trend Direction")
plt.xlabel("Trend Direction")
plt.ylabel("Content Age (Days)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Does Content Age Dictate Trend Direction?

#### 1. Hypothesis

Standard SEO strategy assumes that search engines decay older content in favor of "freshness." If this holds true, pages trending `up` should skew younger (lower `content_age_days`), and pages trending `down` should skew older (higher `content_age_days`).

#### 2. Data Summary

| Trend Direction | Page Count | Median Age (Days) | Mean Age (Days) | Standard Deviation |
| --- | --- | --- | --- | --- |
| **`down` (Declining)** | 16,262 | **216.0** | 236.2 | 127.0 |
| **`flat` (Stagnant)** | 1,152 | **231.0** | 245.9 | 117.5 |
| **`new` (Newly Tracked)** | 2,236 | **279.0** | 238.7 | 100.3 |
| **`stable` (Healthy)** | 5,962 | **300.0** | 295.4 | 139.0 |
| **`up` (Growing)** | 4,388 | **291.5** | 288.5 | 144.5 |

#### 3. Key Observations

* **Growing pages skew older than declining pages:** Pages trending `up` have a median age of 291.5 days—roughly 75 days older than pages actively trending `down` (median 216.0 days). This directly contradicts the freshness assumption.
* **Stable pages have the highest median age:** High-performing, stable pages are the oldest cohort in the dataset (median 300 days), indicating that established authority and intent satisfaction outlast content age.
* **The ~200-day decay clustering:** The high volume of declining pages clustered at a median of 216 days suggests a potential "honeymoon phase" exit. Newer content may receive a temporary ranking boost upon indexing, only to settle into its true, lower organic baseline around month seven.

#### 4. Limitations & Context

* **Survivorship Bias:** The older pages in the `stable` and `up` categories are the winners. The dataset does not capture underperforming older pages that administrators already deleted, redirected, or stopped tracking.
* **Confounding Variables:** Age likely correlates with backlink accumulation and domain authority. These factors, rather than age itself, drive the observed performance stability.

**Conclusion:** These directional trends do not prove search engine mechanics. However, they indicate that **content age is not a reliable predictor** of performance decline, and freshness is not a prerequisite for stability.
