## 4. Careful words: what I can and can't claim

```python
claims_allowed = ["observed associations", "directional trends", "decision-support prioritization"]
claims_forbidden = ["causal proof of recovery", "algorithm reverse-engineering", "guaranteed traffic returns"]

print("ESTABLISHED BOUNDARIES:")
print(f"Allowed Claims: {claims_allowed}")
print(f"Forbidden Claims: {claims_forbidden}")
```
